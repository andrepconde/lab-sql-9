```python
query = "SELECT customer_id, COUNT(rental_id) AS Number_of_rentals FROM rentals_may GROUP BY customer_id HAVING COUNT(rental_id)"

rent_may= pd.read_sql_query(query, engine)
rent_may.head()
```

```python
query = "SELECT customer_id, COUNT(rental_id) AS Number_of_rentals FROM rentals_june_right GROUP BY customer_id HAVING COUNT(rental_id)"

rent_june = pd.read_sql_query(query, engine)
rent_june.head()
```
